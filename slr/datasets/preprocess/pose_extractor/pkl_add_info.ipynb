```python
import os
import pandas as pd
import pickle
```

## CSL Daily

```python
anno_file = "/new_home/xzj23/workspace/SLR/data/csl-daily/sentence_label/csl2020ct_v2.pkl"
pkl_file = "/new_home/xzj23/workspace/SLR/data/keypoints/bak/csl-daily-keypoints.pkl"
save_file = "/new_home/xzj23/workspace/SLR/data/keypoints/kps_info/csl-daily-keypoints.pkl"
```

```python
with open(anno_file, 'rb') as f:
    annotations = pickle.load(f)
info = pd.DataFrame(annotations["info"])
info.set_index("name", inplace=True)
```

```python
info
```

```python
with open(pkl_file, "rb") as f:
    keypoints = pickle.load(f)
```

```python
keypoints.keys()
```

```python
for k in keypoints.keys():
    keypoints[k] = {**keypoints[k], **info.loc[k].to_dict()}
```

```python
with open(save_file, "wb") as f:
    pickle.dump(keypoints, f)
```

## Phoenix2014

```python
annotations_dir = "/new_home/xzj23/workspace/SLR/data/phoenix2014/phoenix-2014-multisigner/annotations/manual"
pkl_file = "/new_home/xzj23/workspace/SLR/data/keypoints/bak/phoenix2014-keypoints.pkl"
save_file = "/new_home/xzj23/workspace/SLR/data/keypoints/kps_info/phoenix2014-keypoints.pkl"
```

```python
corpus = {
    "train": pd.read_csv(os.path.join(annotations_dir, 'train.corpus.csv'),
                         sep='|', header=0, index_col='id'),
    "dev": pd.read_csv(os.path.join(annotations_dir, 'dev.corpus.csv'),
                       sep='|', header=0, index_col='id'),
    "test": pd.read_csv(os.path.join(annotations_dir, 'test.corpus.csv'),
                        sep='|', header=0, index_col='id')
}
info = pd.concat([corpus[m].assign(split=m) for m in corpus.keys()], axis=0, ignore_index=False)
```

```python
info
```

```python
with open(pkl_file, "rb") as f:
    keypoints = pickle.load(f)
```

```python
keypoints.keys()
```

```python
for k in keypoints.keys():
    keypoints[k] = {**keypoints[k], **info.loc[k].to_dict()}
```

```python
with open(save_file, "wb") as f:
    pickle.dump(keypoints, f)
```
