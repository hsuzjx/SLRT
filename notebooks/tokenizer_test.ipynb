```python
import os.path

import numpy as np
from spacy.tests.conftest import tokenizer
from spacy.tests.tokenizer.test_tokenizer import test_tokenizer_special_cases_with_affixes_preserve_spacy
from transformers import BertTokenizer
```

```python
gloss_dict = np.load('/new_home/xzj23/workspace/SLR/data/global_files/gloss_dict/phoenix2014_gloss_dict.npy',
                     allow_pickle=True).item()
vocab_file = '/new_home/xzj23/workspace/SLR/data/global_files/gloss_dict/phoenix2014_vocab.txt'

if not os.path.exists(vocab_file):
    vocab = ['[CLS]', '[SEP]', '[PAD]', '[UNK]', '[MASK]']
    for gloss in gloss_dict.keys():
        vocab.append(gloss)
    with open(vocab_file, 'w') as f:
        for word in vocab:
            f.write(word + '\n')
```

```python
tokenizer = BertTokenizer(vocab_file, clean_up_tokenization_spaces=True)
```
