```python
from slr.datasets import Phoenix2014Dataset
import numpy as np
from torch.utils.data.dataloader import DataLoader
```

```python
gloss_dict = np.load('../experiments/global_file/gloss_dict/phoenix2014_gloss_dict.npy', allow_pickle=True).item()
gloss_dict
```

```python
dataset = Phoenix2014Dataset(features_path='../data/phoenix2014/phoenix-2014-multisigner/features/fullFrame-256x256px',
                             annotations_path='../data/phoenix2014/phoenix-2014-multisigner/annotations/manual',
                             gloss_dict=gloss_dict)
dataset
```

```python
dl = DataLoader(dataset, batch_size=2, shuffle=True, num_workers=4, collate_fn=dataset.collate_fn, drop_last=True,
                pin_memory=True)
```
