```python
import os
from transformers import PreTrainedTokenizer

# 定义词汇表字典
vocab = {
    '<si>': 0,
    '<unk>': 1,
    '<pad>': 2,
    '</s>': 3,
    '我': 4,
    '他': 5,
    '你': 6,
    '这': 7,
    '人': 8,
    '不': 9,
    '多': 10
}


class CustomTokenizer(PreTrainedTokenizer):
    def __init__(self, vocab, **kwargs):
        super().__init__(**kwargs)
        self.vocab = vocab
        self.special_tokens = {'<si>': 0, '<unk>': 1, '<pad>': 2, '</s>': 3}
        self.unk_token = '<unk>'
        self.pad_token = '<pad>'

    def _tokenize(self, text):
        # 将文本拆分为单个字符
        return list(text)

    def _convert_token_to_id(self, token):
        return self.vocab.get(token, self.vocab[self.unk_token])

    def _convert_id_to_token(self, index):
        for token, idx in self.vocab.items():
            if idx == index:
                return token
        return self.unk_token

    def convert_tokens_to_string(self, tokens):
        return ''.join(tokens)

    def get_vocab(self):
        # 返回词汇表
        return self.vocab

    def save_vocabulary(self, save_directory, filename_prefix=None):
        # 保存词汇表到文件
        with open(os.path.join(save_directory, filename_prefix + '-vocab.txt'), 'w', encoding='utf-8') as writer:
            for token, index in self.vocab.items():
                writer.write(f"{token}\t{index}\n")
        return (os.path.join(save_directory, filename_prefix + '-vocab.txt'),)

    @classmethod
    def from_pretrained(cls, pretrained_model_name_or_path, *inputs, **kwargs):
        # 从预训练模型路径加载tokenizer
        vocab = {}
        with open(os.path.join(pretrained_model_name_or_path, 'vocab.txt'), 'r', encoding='utf-8') as f:
            for line in f:
                token, index = line.strip().split('\t')
                vocab[token] = int(index)
        return cls(vocab, **kwargs)


# 实例化自定义 Tokenizer
tokenizer = CustomTokenizer(vocab=vocab)

# 测试文本
text = "我他你好"

# 对文本进行编码
input_ids = tokenizer.encode(text, add_special_tokens=True)
print("Input IDs:", input_ids)

# 解码回文本
decoded_text = tokenizer.decode(input_ids)
print("Decoded Text:", decoded_text)

```
